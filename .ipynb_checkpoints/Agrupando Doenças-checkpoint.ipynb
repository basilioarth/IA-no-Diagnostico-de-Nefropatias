Primeira Agrupamento

```python
data_categorica1['Diagnóstico'].value_counts()
```

```python
i = 0

while i < 100:
    if (data_categorica1['Diagnóstico'][i].lower().__contains__('glomerulonefrite')) | (data_categorica1['Diagnóstico'][i].lower().__contains__('gloemrulonefrite')):
        data_categorica1['Diagnóstico'][i] = 'Glomerulonefrite'
        
    elif data_categorica1['Diagnóstico'][i].lower().__contains__('nefropatia'):
        data_categorica1['Diagnóstico'][i] = 'Nefropatia'
        
    elif data_categorica1['Diagnóstico'][i].lower().__contains__('glomerulopatia'):
        data_categorica1['Diagnóstico'][i] = 'Glomerulopatia'
        
    elif data_categorica1['Diagnóstico'][i].lower().__contains__('podocitopatia'):
        data_categorica1['Diagnóstico'][i] = 'Podocitopatia'
        
    elif data_categorica1['Diagnóstico'][i].lower().__contains__('glomeruloesclerose'):
        data_categorica1['Diagnóstico'][i] = 'Glomeruloesclerose'
        
    elif data_categorica1['Diagnóstico'][i].lower().__contains__('nefroesclerose'):
        data_categorica1['Diagnóstico'][i] = 'Nefroesclerose'
    
    i += 1
```

```python
i = 0

while i < 100:
    if (data_categorica2['Diagnóstico'][i].lower().__contains__('glomerulonefrite')) | (data_categorica2['Diagnóstico'][i].lower().__contains__('gloemrulonefrite')):
        data_categorica2['Diagnóstico'][i] = 'Glomerulonefrite'
        
    elif data_categorica2['Diagnóstico'][i].lower().__contains__('nefropatia'):
        data_categorica2['Diagnóstico'][i] = 'Nefropatia'
        
    elif data_categorica2['Diagnóstico'][i].lower().__contains__('glomerulopatia'):
        data_categorica2['Diagnóstico'][i] = 'Glomerulopatia'
        
    elif data_categorica2['Diagnóstico'][i].lower().__contains__('podocitopatia'):
        data_categorica2['Diagnóstico'][i] = 'Podocitopatia'
        
    elif data_categorica2['Diagnóstico'][i].lower().__contains__('glomeruloesclerose'):
        data_categorica2['Diagnóstico'][i] = 'Glomeruloesclerose'
        
    elif data_categorica2['Diagnóstico'][i].lower().__contains__('nefroesclerose'):
        data_categorica2['Diagnóstico'][i] = 'Nefroesclerose'
    
    else: data_categorica2['Diagnóstico'][i] = 'Outras'
    
    i += 1
```
