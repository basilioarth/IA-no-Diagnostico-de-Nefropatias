```python
import cv2 as cv
import numpy as np
import math as math
import pandas as pd
import matplotlib.pyplot as plt
```

Calculando o Histograma da Imagem:

```python
def calc_hist(imagem):
    return cv.calcHist([imagem], [0], None, [256], [0,256])
```

Calculando o Descritor de Suavidade Relativa:

```python
def r(imagem):
    return ( 1 - ( 1/( 1 + (imagem.var()/(255*255)) ) ) )
```

Calculando o Terceiro Momento:

```python
def u3(imagem, hist_00):
    dimensao = imagem.shape[0] * imagem.shape[1]
    i = 0
    u3 = 0
    m = imagem.mean()

    while i < len(hist_00):
        # hist_00[i][0]/dimensao informa a probabilidade de ocorrência dessa intensidade na imagem
        atual = ( ((i - m) ** 3) * (hist_00[i][0]/dimensao) )
        u3 = u3 + atual
        i = i + 1
        
    return u3
```

Calculando a Uniformidade:

```python
def uniformidade(imagem, hist_00):
    i = 0
    uniformidade = 0
    dimensao = imagem.shape[0] * imagem.shape[1]

    while i < len(hist_00):
        # hist_00[i][0]/dimensao informa a probabilidade de ocorrência dessa intensidade na imagem
        atual = ( (hist_00[i][0]/dimensao) ** 2 )
        uniformidade = uniformidade + atual
        i = i + 1
        
    return uniformidade
```

Calculando a Entropia:

```python
def entropia(imagem, hist_00):
    i = 0
    entropy = 0
    dimensao = imagem.shape[0] * imagem.shape[1]
    
    while i < len(hist_00):
        # hist_00[i][0]/dimensao informa a probabilidade de ocorrência dessa intensidade na imagem
        p = (hist_00[i][0]/dimensao)
        if p > 0:
            atual = ( (p * math.log2(p)) * (-1) )
            #print(atual)
            entropy = entropy + atual
        i = i + 1
    
    return entropy
```

Inicializando Novas Colunas:

```python
base_limpa1['Descritor de Suavidade Relativa'] = -1.000
base_limpa1['Terceiro Momento'] = -1.000
base_limpa1['Uniformidade'] = -1.000
base_limpa1['Entropia'] = -1.000
base_limpa1
```

Retirando Atributos da Imagem e Inserindo na Base de Dados:

```python
contador = 1

while contador < 72:
    path = "C:\\Users\\LENOVO\\Desktop\\Arthur\\LINA\\PIBIC 2019-2020 (01.08.2019 a 31.07.2020)\\Imagens Renais\\" + str(contador) + ".jpg"
    imagem = cv.imread(path, 0)
    
    # Calculando o Histograma da Imagem:
    hist_00 = calc_hist(imagem)
    
    # Calculando o Descritor de Suavidade Relativa da Imagem:
    base_limpa1['Descritor de Suavidade Relativa'][contador -1] = round(r(imagem), 3)
    
    # Calculando o Terceiro Momento da Imagem:
    base_limpa1['Terceiro Momento'][contador -1] = round(u3(imagem, hist_00), 3)
    
    # Calculando a Uniformidade da Imagem:
    base_limpa1['Uniformidade'][contador -1] = round(uniformidade(imagem, hist_00), 3)
    
    # Calculando a Entropia da Imagem:
    base_limpa1['Entropia'][contador -1] = round(entropia(imagem, hist_00), 3)
    
    contador += 1
```

Retirando Atributos da Imagem e Inserindo na Base de Dados:

```python
contador = 1

while contador < 72:
    path = "C:\\Users\\LENOVO\\Desktop\\Arthur\\LINA\\PIBIC 2019-2020 (01.08.2019 a 31.07.2020)\\Imagens Renais\\" + str(contador) + ".jpg"
    imagem = cv.imread(path, 0)
    
    # Calculando a Média da Imagem:
    media = imagem.mean()
    
    # Calculando o Desvio Padrão da Imagem:
    desvio = imagem.std()
 
    # Calculando o Histograma da Imagem:
    hist_00 = calc_hist(imagem)
    
    # Calculando o Descritor de Suavidade Relativa da Imagem:
    suavidade_relativa = r(imagem)
    
    # Calculando o Terceiro Momento da Imagem:
    terceiro_momento = u3(imagem, hist_00)
    
    # Calculando a Uniformidade da Imagem:
    unifor = uniformidade(imagem, hist_00)
   
    # Calculando a Entropia da Imagem:
    entro = entropia(imagem, hist_00)
    
    # Classificando a Textura:
    if ((media <= 83) & (desvio <= 12) & (suavidade_relativa <= 0.002) & (unifor >= 0.026) & (entro <= 5.434)):
        base_limpa2['Textura'][contador - 1] = 'Suave'
        
    elif((media >= 144) & (desvio >= 75) & (suavidade_relativa >= 0.079) & (unifor <= 0.005) & (entro >= 7.783)):
        base_limpa2['Textura'][contador - 1] = 'Rugosa'
        
    else:
        base_limpa2['Textura'][contador - 1] = 'Regular'
    
    contador += 1
```
