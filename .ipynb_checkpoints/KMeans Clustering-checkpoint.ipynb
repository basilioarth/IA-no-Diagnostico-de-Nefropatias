# Base Original

```python
base_original = base_numerica_1atributo.drop(['Textura_Inexistente', 'Textura_Regular', 'Textura_Suave'], axis = 1)
```

```python
base_original.head()
```

Selecioando os 71 primeiros registros:

```python
base_original = base_original[0:71]
base_original
```

Descobrindo o valor inicial para K

```python
inertia = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, random_state=1234)
    kmeans.fit(base_original.drop(['Diagnóstico'], axis = 1))
    inertia.append((i,kmeans.inertia_,))
```

```python
plt.plot([w[0] for w in inertia],[w[1] for w in inertia], marker="X")
```

```python
kmeans = KMeans(n_clusters = 8, init = 'k-means++', n_init = 10, random_state=1234)
```

```python
kmeans.fit(base_original.drop(['Diagnóstico'], axis = 1))
```

Soma dos Erros Quadráticos:

```python
kmeans.inertia_
```

Silhouette Coefficient:

```python
s = silhouette_score(base_original.drop(['Diagnóstico'], axis = 1).values, kmeans.labels_, metric='sqeuclidean')
print(s)
```

Calinski-Harabasz Index:

```python
metrics.calinski_harabaz_score(base_original.drop(['Diagnóstico'], axis = 1), kmeans.labels_)
```

Davies-Bouldin Index:

```python
davies_bouldin_score(base_original.drop(['Diagnóstico'], axis = 1), kmeans.labels_)
```

# Hierarchical clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
clustering.fit(base_numerica_1atributo.drop("Diagnóstico", axis = 1))
```

```python
s = silhouette_score(base_numerica_4atributos.drop(['Diagnóstico'], axis = 1).values, clustering.labels_, metric='sqeuclidean')
print(s)
```

Calinski-Harabasz Index:

```python
metrics.calinski_harabaz_score(base_numerica_4atributos.drop(['Diagnóstico'], axis = 1), clustering.labels_)
```

Davies-Bouldin Index:

```python
davies_bouldin_score(base_numerica_4atributos.drop(['Diagnóstico'], axis = 1), clustering.labels_)
```
