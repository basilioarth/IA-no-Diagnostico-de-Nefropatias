```python
for i in range(0, 100):
    if ((data['Diagnóstico'][i] == 'GLOMERULONEFRITE LÚPICA DIFUSA E GLOBAL- CLASSE IV')                       | 
        (data['Diagnóstico'][i] == 'Glomerulonefrite membranoproliferativa tipo II(Doença de depósito denso)') |
        (data['Diagnóstico'][i] == 'Glomerulonefrite membranoproliferativa tipo I')                            |
        (data['Diagnóstico'][i] == 'Glomerulonefrite membranoproliferativa tipo III')                          |
        (data['Diagnóstico'][i] == 'GLOMERULONEFRITE LÚPICA FOCAL E SEGMENTAR - CLASSE III')                   |
        (data['Diagnóstico'][i] == 'Glomerulonefrite membranoproliferativa')                                   |
        (data['Diagnóstico'][i] == 'Glomerulonefrite crescêntica')                                             |
        (data['Diagnóstico'][i] == 'GLOMERULONEFRITE LÚPICA SEGMENTAR E FOCAL - CLASSE III')                   |
        (data['Diagnóstico'][i] == 'Glomerulonefrite difusa aguda (GNDA)')                                     |
        (data['Diagnóstico'][i] == 'GLOMERULONEFRITE LÚPICA DIFUSA E GLOBAL - CLASSE IV')                      |
        (data['Diagnóstico'][i] == 'GLOMERULONEFRITE LÚPICA DIFUSA E SEGMENTAR ´CLASSE IV')                    |
        (data['Diagnóstico'][i] == 'GLOMERULONEFRITE LÚPICA MESANGIAL PROLIFERATIVA – CLASSE II')              |
        (data['Diagnóstico'][i] == 'GLOMERULONEFRITE LÚPICA MEMBRANOSA- CLASSE V')                             |
        (data['Diagnóstico'][i] == 'GLOMERULONEFRITE LÚPICA FOCAL E GLOBAL - CLASSE III')                      |
        (data['Diagnóstico'][i] == 'GLOMERULONEFRITE LÚPICA SEGMENTAR E FOCAL – CLASSE III')                   |
        (data['Diagnóstico'][i] == 'GLOMERULONEFRITE LÚPICA SEGMENTAR E FOCAL – CLASSE III (A) E ASSOCIAÇÃO COM MEMBRANOSA (CLASSE V)')):
        
        data['Diagnóstico'][i] = 'Glomerulonefrite'
        
    elif ((data['Diagnóstico'][i] == 'Nefropatia diabética') | (data['Diagnóstico'][i] == 'Nefropatia por IgA') | 
          (data['Diagnóstico'][i] == 'Nefropatia por IgA (doença de Berger)')):
        
        data['Diagnóstico'][i] = 'Nefropatia'
    
    elif((data['Diagnóstico'][i] == 'Glomerulopatia membranosa secundária') | (data['Diagnóstico'][i] == 'Glomerulopatia membranosa primária') | 
         (data['Diagnóstico'][i] == 'Glomerulopatia membranosa') | (data['Diagnóstico'][i] == 'Glomerulopatia por fibronectina')):
        
        data['Diagnóstico'][i] = 'Glomerulopatia'
        
    elif((data['Diagnóstico'][i] == 'Doença de Lesões mínimas') | (data['Diagnóstico'][i] == 'Glomeruloesclerose segmentar e focal') | 
         (data['Diagnóstico'][i] == 'Podocitopatias (Lesões Mínimas X Glomeruloesclerose Segmentar e Focal não amostrada)') | 
         (data['Diagnóstico'][i] == 'Podocitopatia: Glomeruloesclerose segmentar e Focal não amostrada') | 
         (data['Diagnóstico'][i] == 'Glomeruloesclerose segmentar e focal variante colapsante')):
        
        data['Diagnóstico'][i] = 'Podocitopatia'
        
    else:
        data['Diagnóstico'][i] = 'Outros'
        
```
